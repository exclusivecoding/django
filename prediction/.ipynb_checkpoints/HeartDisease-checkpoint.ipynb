```python
import pandas as pd
from collections import Counter
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression 
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import seaborn as sns
import matplotlib.pyplot as plt 
```

**Model Development and Prediction**

```python
X = data.drop(['target'],axis=1)
y = data['target']
#split the data in train and test
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#===================================Model : Logistic Regression=====================
scaler = StandardScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)

LR_model = LogisticRegression(max_iter=1000)
LR_model.fit(X_train_scale,y_train)

pred_LR = LR_model.predict(X_test_scale)
accu_LR = accuracy_score(y_test,pred_LR)
print(f'Accuracy Score : {accu_LR*100:.2f}%')
cnf_LR = confusion_matrix(y_test,pred_LR)
print(cnf_LR)
confusion_matrix_plot(y_test,pred_LR,'Linear Regression')
clr_LR = classification_report(y_test,pred_LR)
print(clr_LR)
```

```python
#Plot for Confusion Matrix 
def confusion_matrix_plot(y_test,pred_y,model):
    cmp = confusion_matrix(y_test,pred_y)
    plt.figure(figsize=(6,4))
    sns.heatmap(cmp,annot=True,fmt='d',cmap='Blues',cbar=False)
    plt.title(f'Confusion Matrix for {model}')
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
    
```

```python
#====================================Model : Random Forest==============================================================

RF_model = RandomForestClassifier(random_state=42)
RF_model.fit(X_train,y_train)

pred_RF = RF_model.predict(X_test)
accu_RF = accuracy_score(y_test,pred_RF)
print(f'Accuracy Score : {accu_RF*100:.2f}%')
cnf_RF = confusion_matrix(y_test,pred_RF)
print(cnf_RF)
confusion_matrix_plot(y_test,pred_RF,'Random Forest')
clr_RF = classification_report(y_test,pred_RF)
print(clr_RF)

#===============================================Explanation================================================================
#Here TP:192, FP:13, FN:11, TN:141
#False Positive : 13 - The model incorrectly predicted class 1 (disease) when it was actually class 0 (normal).
#False Negative : 11 - The model incorrectly predicted class 0 (normal) when it was actually class 1 (disease).

#Precision for class 1 (disease) : 0.94
#Precision measures the proportion of true positives out of the total predicted positives. for class 1: 192/(192+13) = ~0.937

#Recall for class 1 (disease) : 0.95
#Recall measures the proportion of true positives out of the total actual positives. for class 1 : 192/(192+11) = ~0.946

#Precision for class 0 (normal) : 0.93
#Precision measures the proportion of true Negative out of the total predicted Negative. for class 0: 141/(141+11) = ~0.928

#Recall for class 0 (normal) : 0.92
#Recall measures the proportion of true Negative out of the total actual Negative. for class 0 : 141/(141+13) = ~0.916
```

```python
#====================================Model : XGB ==============================================================

XGB_model = XGBClassifier(learning_rate=0.2,subsample=0.9)
XGB_model.fit(X_train,y_train)

pred_XGB = XGB_model.predict(X_test)
accu_XGB = accuracy_score(y_test,pred_XGB)
print(f'Accuracy Score : {accu_XGB*100:.2f}%')
cnf_XGB = confusion_matrix(y_test,pred_XGB)
print(cnf_XGB)
confusion_matrix_plot(y_test,pred_XGB,'eXtreme Gradient Boosting')
clr_XGB = classification_report(y_test,pred_XGB)
print(clr_XGB)
```

```python
#=======================================predict on user input======================================
#For reference : Help to choose the value
# 1) Age 

# 2) sex
# 0 = female,
# 1 = male 
 
# 3) chest pain type 
# 1 = typical angina,
# 2 = atypical angina,
# 3 = non-anginal pain,
# 4 = asymptomatic Nominal

# 4) resting blood pressure : resting bp s 

# 5) cholesterol 

# 6) fasting blood sugar
# 1 = sugar > 120mg/dL
# 0 = sugar < 120mg/dL 

# 7) resting electrocardiogram : results resting ecg
# 0 = normal,
# 1 = ST-T wave abnormality (T wave inversions
# and/or ST elevation/depression of > 0.05 mV),
# 2 = Probable or Definite Left Ventricular
# hypertrophy by Estes criteria Nominal

# 8) maximum heart rate

# 9) exercise induced angina : exercise angina
# 0 = no,
# 1 = yes 

# 10) oldpeak =ST : oldpeak 

# 11) the slope of the peak exercise ST segment : ST slope
# 1 = upward
# 2 = flat,
# 3 = downward Nominal

# class target
# 0 = Normal,
# 1 = Heart Disease Binary


def user_input():
    print("Enter the Following Details for Prediction:")
    
    user_inp = []                                                   
    for i in X_train.columns:
        value = float(input(f'Enter the value for {i}: '))              
        user_inp.append(value)                                        #Append the value in variable 
    return user_inp

def main():
    while True:
        user = int(input("Do you want to continue or Exit for prediction? Press 1 or 0: (continue:1,Exit:0)"))
        if user == 0:
                   print("Exit...")
                   break
        elif user == 1:
                   user_inp = user_input()
                   pred_user = XGB_model.predict([user_inp])
                   print("="*100)
                   print(f"********Patient has {'Heart Disease' if pred_user[0] == 1 else 'Normal'} Condition********")
        else:
                   print("*******Invalid Input*********")

if __name__ == "__main__":
    main()
```
